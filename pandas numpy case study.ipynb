SALES AND PROFIT BY CATEGORY AND SUBCATEGORY

```python
#SALES
x = df.groupby(['Category','Sub-Category'])['Sales'].sum()
x
```

```python
fig=plt.figure()
ax0=fig.add_subplot(2,2,1)
ax1=fig.add_subplot(2,2,2)
ax2=fig.add_subplot(2,2,3)
ax3=fig.add_subplot(2,2,4)

x.sort_values(ascending=True).plot(kind='barh',figsize=(8,8),ax=ax0)
ax0.set_title("All Sub-Categories")
ax0.set_xlabel("Sub-Categories")
ax0.set_ylabel("Sales")

x['Furniture'].sort_values(ascending=True).plot(kind='barh',figsize=(12,10),ax=ax1)
ax1.set_title("Furniture")
ax1.set_xlabel("Sub-Categories")
ax1.set_ylabel("Sales")

x['Office Supplies'].sort_values(ascending=True).plot(kind='barh',figsize=(8,8),ax=ax2)
ax2.set_title("Office Supplies")
ax2.set_xlabel("Sub-Categories")
ax2.set_ylabel("Sales")

x['Technology'].sort_values(ascending=True).plot(kind='barh',figsize=(12,10),ax=ax3)
ax3.set_title("Technology")
ax3.set_xlabel("Sub-Categories")
ax3.set_ylabel("Sales")
```

1.MAXIMUM SALES : PHONES FROM TECHNOLOGY, CHAIRS FROM FURNITURE , STORAGE FROM OFFICE SUPPLIES

2.MINIMUM SALES: FASTENERS , LABELS AND ENVELOPES FROM OFFICE SUPPLIES

```python
#Profit
y = df.groupby(['Category','Sub-Category'])['Profit'].sum()
y
```

```python
fig=plt.figure()
ax0=fig.add_subplot(2,2,1)
ax1=fig.add_subplot(2,2,2)
ax2=fig.add_subplot(2,2,3)
ax3=fig.add_subplot(2,2,4)

y.sort_values(ascending=True).plot(kind='barh',figsize=(8,8),ax=ax0)
ax0.set_title("All Sub-Categories")
ax0.set_xlabel("Sub-Categories")
ax0.set_ylabel("Profit")

y['Furniture'].sort_values(ascending=True).plot(kind='barh',figsize=(12,10),ax=ax1)
ax1.set_title("Furniture")
ax1.set_xlabel("Sub-Categories")
ax1.set_ylabel("Profit")

y['Office Supplies'].sort_values(ascending=True).plot(kind='barh',figsize=(8,8),ax=ax2)
ax2.set_title("Office Supplies")
ax2.set_xlabel("Sub-Categories")
ax2.set_ylabel("Profit")

y['Technology'].sort_values(ascending=True).plot(kind='barh',figsize=(12,10),ax=ax3)
ax3.set_title("Technology")
ax3.set_xlabel("Sub-Categories")
ax3.set_ylabel("Profit")
```

1.MAXIMUM PROFIT : COPIERS , PHONES AND ACCESSORIES IN TECHNOLOGY

2.LOSS : SUPPLIES(OFFICE SUPPLIES),BOOKCASE AND TABLES (FURNITURE)
